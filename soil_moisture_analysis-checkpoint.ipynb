```python
#quick linear regression
import matplotlib.pyplot as plt
from scipy import stats


x = sm_data['Snow depth (in)']
y = sm_data['Soil Moisture(%)']
mask = ~np.isnan(x) & ~np.isnan(y) #mask NaN values

slope, intercept, r_value, p_value, std_err = stats.linregress(x[mask], y[mask]) #outputs slope, intercept

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
mymodel
display(r_value)
display(std_err)


plt.scatter(x, y)
plt.plot(x, mymodel)
plt.legend([f'R^2 = {r_value:.3f}']) #add r^2
plt.show()

fig = plt.figure(figsize=(20,10), facecolor='white')
ax1 = fig.add_subplot(1,1,1)

ax1.plot(x, mymodel, color = 'tab:blue', label = 'Regression, R$^{2}$ = '+str(r_value)[:6])
ax1.scatter(sm_data['Snow depth (in)'], sm_data['Soil Moisture(%)'], color = 'tab:blue', label = 'Soil Moisture')
ax1.tick_params(axis='x', labelrotation=45)
ax1.set_xlim([(-0.5), (6.5)]) #set x axis limits
ax1.set_ylim([0, 55]) #set y axis limits
ax1.tick_params(axis='y', colors='tab:blue', labelsize=20)
ax1.set_ylabel('Soil Moisture (%)', size=axis_label_size, color = 'tab:blue')
ax1.set_xlabel('Snow Depth (in)', size=axis_label_size, color = 'black')
ax1.tick_params(axis='x', colors='black', labelsize=20)
ax1.legend(prop={'size': 20}, loc = 'upper right')
#ax1.set_yticks(np.arange(0, 25.1, 5)) #set y axis ticks


fig.savefig('Snow Depth vs Soil Moisture', dpi=300, bbox_inches='tight')
```
